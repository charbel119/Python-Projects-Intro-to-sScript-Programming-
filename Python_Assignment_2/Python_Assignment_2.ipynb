```python
# Submitted by: 








# This fuction perfome data file handling (open and read the file contained data) when called
# Parameter: file = the name of the file
def read_file(file):
    
    # Statements to open the desired file
    # Then, reading the opened file
    with open(file) as f:
        seq= f.read()       
    
    # Statement to terminate the function by returning the defined object by this function
    return seq








# This function create a directory to put the generated file from this script
# Parameter: folder = the name of the folder that has to be created
def directory(folder):
    
    # Statements to create a folder in the present location with the python module os
    # That statements can check either the folder was created or not created.
    # If the folder was not created, than create one with user defined folder name
    # If the folder was created, than release a message. 
    import os                
    if os.path.exists(folder)==True:      
        template = "Folder {} is already existed".format(folder)
        return template
    else:                                    
        dir_ = os.mkdir(folder)
        return dir_

    

    
    
    
# Calling the function directory(), so that the folder "output" will be created    
directory("output") 

```

```python
# This function remove newline character from the string object
# Parameter: file = the name of the file from which the newline has to be removed
def remove_newline(file):
    
    
    # Remove the newline character is worked by splitting the line on the basis of '\n' and rejoining the splitted sequence
    # Calling the function read_file() is needed to specify the file from which the newline character will be removed.
    seq = read_file(file)
    
    
    seq = seq.split('\n')
    
    
    seq = ''.join(i for i in seq)
    
    
    return seq



```

```python
# This function extract exons from the gene sequence and create mRNA sequence
# Parameters: file1 = gene sequence file-name, file2 = exon postions file-name

def forming_mRNA(file1, file2):
    
    
    # Statements remove newline character from the file1
    # Read exons positions file and splitted into multiple line
    seq = remove_newline(file1)
    
    exon_positions = read_file(file2)
    
    splitted_exon_positions = exon_positions.split("\n")
    
    
    # Statements perform list comprehension to extract the exons from the gene sequence for the specific range of indices to a string object
    # That specific range of indices are specified in the lines of exon positions file 
    # Each set of element is adding to the string in each step of the for loop
    exon_specific_nucleotides = ""
    for i in splitted_exon_positions:
        start,end = i.split()
        start,end = int(start),int(end)
        exon_specific_nucleotides += seq[start:end]
        
    
    # Statements to form the mRNA by replacing the letter "T" with "U"
    mRNA = exon_specific_nucleotides
    mRNA = mRNA.replace("T","U")
    
    
    
    
    return mRNA
        
    
```

```python
# This function create protein from mRNA sequence
# Parameters: file1 = the file-name of gene sequence, file2 = the file-name of exon-positions and
# file3 = the filename of the code
def creating_protein(file1 = "gene_sequence.txt",file2 = "exon_positions.txt", file3 = "code.txt"):
    
    
    
    # Statements to call the functions, so that the the information of that functions can be used in this function
    # Parameters are used as the arguments of the that functions
    # mRNA sequence and code are defined by calling the functions
    # The code is splitted into multiple lines 
    mRNA = forming_mRNA(file1,file2)
    
    code = read_file(file3)
    code_ = code.split('\n')
    
    
    
    # Statements perform list comprehension to fillup a dictionary
    # The keywords of the dictionary are the element of 1st column of the code (inside the code, each column is separated by the tab)
    # Each set of elements is adding to the dictionary in each step of the for loop
    cd = code_.copy()
    
    code_ = {}
    
    for i in cd:
        j = i.split()
        
        code_[j[0]] = j[1]
        
    
    
    # Statements perform the list comprehension to form the string of amino acids
    # Amino acid is formed by mapping every 3 nucleotides to a corresponded amino acid
    # Each triplets of nucleotides in the form of mRNA is added with a string in each loop
    # The loop started, when one start codon is found in the mRNA sequence
    # The loop is stopped or breaken just before one triplets mapping into the amino acid 'X'
    # 'X' is the amino acid symbol for end codon, thus loop breaks, when one of the three end codon is met
    start_constant =  mRNA.find("AUG")
    length = int(len(mRNA[start_constant:])/3)
    
    protein = ""
    for i in range(length):
        start = start_constant+i*3
        stop = start+3
        value_code = code_[mRNA[start:stop]]
        if value_code=='X':
            break
            
        protein += value_code
        


    
    # Statements perform list comprehension per 80 characters of the protein string
    # A newline is created after each 80 characters of the protein string
    # Each 80 characters and a newline character is added with a list at each loop
    # When the elements of the list joined into single string, than line has created by the inserted newline character after each 80 characters
    lines = []
    for i in range(0,len(protein),80):
        line = protein[i:i+80]+'\n'
        lines.append(line)
    
    
    protein  = ''.join(i for i in lines) 
    
    
    
    
        
        

    return protein


# Calling the function creating_protein() to get the output of that function
creating_protein()
```

```python
# This function save the protein sequence as the text file format into the 'output' directory
# Parameters: file1 = the file-name of gene sequence, file2 = the file-name of exon-positions and
# file3 = the filename of the code
def Output_protein_sequence(file1= "gene_sequence.txt",file2 = "exon_positions.txt", file3 = "code.txt"):
    
    
    # Calling the functions, so that the information of that functions can be transferred into this function 
    protein = creating_protein(file1,file2,file3)
    
    directory("output")
    
    
    
    # Saving the protein sequence as the file format
    # The folder 'output' was created and checked if the folder is existed, in the time of calling the function directory()
    output = open("output/Protein_sequence.txt", "w")
    
    output.write(protein)
    
    output.close()
    

# Calling this function is finally saving the protein sequence    
Output_protein_sequence()    
```

```python
# This function can determine the protein structure, length of protein and conclude if mutation happend in the position 30049, T>A
# Parameters: file1 = the file-name of gene sequence, file2 = the file-name of exon-positions and
# file3 = the filename of the code, position = the location where the point mutation happend in gene
# point_mutation = the letter which has to be replaced by the original letter in the time of mutation
def mutation(file1= "gene_sequence.txt",file2 = "exon_positions.txt", file3 = "code.txt", position = 30049, point_mutation = "A"):
    
    
    # To create a mutated gene sequence, gene sequence file is read and the letter 'T' is replaced by the letter 'A'
    # If letter 'T' isn't in the position of point-mutation, than this function will be terminated with an error message
    gene_seq = remove_newline(file1)
    
    if gene_seq[position]=="T":
        mutated_gene_seq = gene_seq[:position]+point_mutation+ gene_seq[position+1:]
    else:
        return "Error, nucleotide must be T in the position"
        
    # Mutated gene sequence is separated by the lines and each line is 80 character long    
    lines = []
    for i in range(0,len(mutated_gene_seq),80):
        line = mutated_gene_seq[i:i+80]+'\n'
        lines.append(line)
    
    
    mutated_gene_seq  = ''.join(i for i in lines)     
        
        
    # Mutated gene is saving into the directory output
    directory("output")
    mut_gene = open("output/Mutated_gene_sequence.txt", "w")
    
    mut_gene.write(mutated_gene_seq)
    
    mut_gene.close()
    
    
    # By the following statements, protein is created for the mutated gene while calling the function mutated_protein
    # The filename of mutated protein sequence is passing as the argument of the function creating_protein, so that, protein is creating for the mutated gene
    file = "output/Mutated_gene_sequence.txt"
    
    mutated_protein = creating_protein(file,file2,file3)
    
    
    
    # Saving the mutated protein sequence
    m_protein = open("output/Mutated_Protein_sequence.txt", "w")
    
    m_protein.write(mutated_protein)
    
    m_protein.close()
    
    
    # Read the normal protein and mutated protein file to compare the length between the normal and mutated protein 
    normal_protein = remove_newline("output/Protein_sequence.txt")
    
    mutated_protein = remove_newline("output/Mutated_Protein_sequence.txt")
    
    difference = len(normal_protein)-len(mutated_protein)
    
    
    # Create the conclusion template which is depended on the value of difference and similarity or dissimalarity between the normal and mutated protein
    if difference>0:
        template1 = "(1) Protein has terminated upstream to the normal ending position due to the mutation and in the   place of termination, a stop codon has appeared. Thus, protein length has reduced and a 'Nonsense' mutation has   taken place."
    elif difference == 0:
        template1 = "(1) The length of mutated Protein is equal to the non-mutated protein."
        
        
    if normal_protein[:len(mutated_protein)] == mutated_protein:
        if difference>0:
            template2 = " (2) Mutation hasn't changed the amino acid structure, unless the stop codon has appeared."
        else:
            template2 = " (2) Mutation hasn't changed the amino acid structure. Thus a 'Silent' mutation has happened"
    else:
        template2 = " (2) Mutation has changed the amino acid structure."
        
        
    template = template1+template2    
        
        
       
    
    # Putting the all statistics of the mutation and conclusion into a dictionary object
    report = {
              "Name_of_the_disease":"Congenital Lactase Deficiency",
              
              "Normal_protein_length": len(normal_protein),
              
              "Mutated_protein_length": len(mutated_protein),
              
              "Difference_normal_minus_mutated_protein_length": difference,
              
              "Conclusion": template
             }
    
    # When function will be called, then that dictionary will be printed
    return report





# Calling the mutation function
mutation()
    
```
